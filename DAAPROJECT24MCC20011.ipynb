```python
import random
import time
import matplotlib.pyplot as plt
# Merge Sort function
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

# Quick Sort function
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

# Generate random list of 1000 items (0-100)
arr = [random.randint(0, 100) for _ in range(1000)]

# Display original array (showing only first 10 for brevity)
print("Original array (first 10 items):", arr[:10])

# Measure time for Merge Sort
merge_arr = arr.copy()
start_time = time.time()
merge_sort(merge_arr)
merge_sort_time = time.time() - start_time

# Measure time for Quick Sort
quick_arr = arr.copy()
start_time = time.time()
quick_sort(quick_arr, 0, len(quick_arr) - 1)
quick_sort_time = time.time() - start_time



# Plotting the results
plt.figure(figsize=(10, 6))
plt.bar(['Merge Sort', 'Quick Sort'], [merge_sort_time, quick_sort_time], color=['blue', 'orange'])
plt.ylabel('Execution Time (seconds)')
plt.title('Merge Sort vs Quick Sort Performance with 1000 Random Items (0-100)')
plt.grid(axis='y')
plt.show()
```
